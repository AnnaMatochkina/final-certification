```python
import pandas as pd
import joblib

# Загружаем модель
model = joblib.load('best_model_rfc.joblib')

feature_names = [
    'Нематериальные активы 1110', 'Результаты исследований и разработок 1120', ' Нематериальные поисковые активы 1130',
    'Материальные поисковые активы 1140', 'Основные средства 1150', 'Доходные вложения в материальные ценности 1160',
    'Финансовые вложения 1170', 'Отложенные налоговые активы 1180', 'Прочие внеоборотные активы 1190',
    'Запасы 1210', 'НДС по приобретенным ценностям 1220', 'Дебиторская задолженность 1230', 'Финансовые вложения 1240',
    'Денежные средства и эквиваленты 1250', 'Прочие оборотные активы 1260',
    'Уставный капитал 1310', 'Собственные акции 1320', 'Переоценка внеоборотных активов 1340', 'Добавочный капитал 1350',
    'Резервный капитал 1360', 'Нераспределенная прибыль 1370',
    'Заемные средства 1410', 'Отложенные налоговые обязательства 1420', 'Оценочные обязательства 1430', 'Прочие обязательства 1450',
    'Заемные средства 1510', 'Кредиторская задолженность 1520', 'Доходы будущих периодов 1530', 'Оценочные обязательства 1540', 'Прочие обязательства 1550',
    'Валовая прибыль (убыток) 2100', 'Прибыль (убыток) от продаж 2200', 'Прибыль (убыток) до налогооблажения 2300', 'Чистая прибыль (убыток) 2400'
]

# Инициализируем список для значений
input_values = []

# Ввод данных по строкам
print('Введите значения строк бухгалтерской отчетности:')
for i, feature in enumerate(feature_names, 1):
  while True:
    try:
      value = float(input(f'{i}. {feature}: '))
      input_values.append(value)
      break
    except ValueError:
      print('Ошибка: введите числовое значение.')

# Проверяем, что введено ровно 34 значения
if len(input_values) != 34:
  print(f'Ошибка: ожидалось 34 значения, получено {len(input_values)}')
else:
  # Создаем DataFrame с одной строкой
  df = pd.DataFrame([input_values], columns=feature_names)
  # Предсказание
  df_np = df.values  # преобразуем DataFrame в numpy array
  prediction = model.predict(df_np)[0]
  proba_array = model.predict_proba(df_np)[0]

  # Вывод результата и рекомендации
  if prediction == 1:
    print(f'\nПрогноз: Организация финансово устойчива (вероятность класса 0 - {proba_array[0]:.2f}, вероятность класса 1 - {proba_array[1]:.2f})')
  else:
    print(f'\nПрогноз: Организация финансово неустойчива (вероятность класса 0 - {proba_array[0]:.2f}, вероятность класса 1 - {proba_array[1]:.2f})')
    print('Рекомендация: Рекомендуется провести дополнительный анализ.')
```
