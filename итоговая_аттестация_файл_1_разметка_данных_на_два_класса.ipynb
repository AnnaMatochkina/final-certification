---
**Модель Шеремета А.Д., Сайфулина Р.С., Негашева Е.В., Кадыкова Г.Г.**

1.  Коэффициент обеспеченности собственными средствами
2.	Коэффициент текущей ликвидности
3.	Коэффициент интенсивности оборота авансируемого капитала
4.	Коэффициент менеджмента
5.	Рентабельность собственного капитала

$$
R = 2K_1 + 0.1K_2 + 0.08K_3 + 0.45K_4 + K_5
$$

Если значение показателя R < 1, то вероятность банкротства организации высокая, если R > 1, то вероятность банкротства – низкая.

```python
df_ShSNK = pd.DataFrame({
    'ID': df_report_not_identified['ID'],
    'own_funds_ratio': (df_report_not_identified['1300'] - df_report_not_identified['1100']) / df_report_not_identified['1200'],
    'cur_ratio': df_report_not_identified['1200'] / (df_report_not_identified['1510'] + df_report_not_identified['1520'] + df_report_not_identified['1550']),
    'turnover_int_advanced_cap_ratio': df_report_not_identified['2110'] / (df_report_not_identified['1100'] + df_report_not_identified['1200']),
    'rate_of_return': df_report_not_identified['2400'] / df_report_not_identified['2110'],
    'ROE_ratio': df_report_not_identified['2400'] / df_report_not_identified['1300'],
})
```

```python
df_ShSNK['R'] = 2 * df_ShSNK['own_funds_ratio'] + 0.1 * df_ShSNK['cur_ratio'] + 0.08 * df_ShSNK['turnover_int_advanced_cap_ratio'] + 0.45 * df_ShSNK['rate_of_return'] + df_ShSNK['ROE_ratio']
```

```python
df_ShSNK['total_ShSNK'] = np.where(df_ShSNK['R'] <= 1, 'unstable', 'not identified')
df_ShSNK['total_ShSNK'] = np.where(df_ShSNK['R'] > 1, 'stable', df_ShSNK['total_ShSNK'])
```

```python
df_ShSNK['total_ShSNK'].value_counts()
```

```python
df_ShSNK_total = df_ShSNK[['ID', 'total_ShSNK']]
```

Еще одной методикой на основе получения интегральной оценки является четырехфакторная **модель Постюшкова А.В.**

$$
R = 0.125*K_1+2.5*K_2+0.4*K_3+1.25*K_4
$$

1.	Коэффициент текущей ликвидности
2.	Коэффициент обеспеченности собственными средствами
3.	Коэффициент оборачиваемости собственного капитала
4.	Рентабельность собственного капитала

Оценку риска банкротства следует проводить по следующей схеме: если R  < 0,99, то вероятность банкротства высокая, если R > 1, то вероятность банкротства низкая.

```python
df_Postushkov = pd.DataFrame({
    'ID': df_report_not_identified['ID'],
    'cur_ratio': df_report_not_identified['1200'] / (df_report_not_identified['1510'] + df_report_not_identified['1520'] + df_report_not_identified['1550']),
    'own_funds_ratio': (df_report_not_identified['1300'] - df_report_not_identified['1100']) / df_report_not_identified['1200'],
    'equity_turnover_ratio': df_report_not_identified['2110'] / df_report_not_identified['1300'],
    'ROE_ratio': df_report_not_identified['2400'] / df_report_not_identified['1300']
})
```

```python
df_Postushkov['R'] = 0.125 * df_Postushkov['cur_ratio'] + 2.5 * df_Postushkov['own_funds_ratio'] + 0.4 * df_Postushkov['equity_turnover_ratio'] + 1.25 * df_Postushkov['ROE_ratio']
```

```python
df_Postushkov['total_postushkov'] = np.where(df_Postushkov['R'] <= 1, 'unstable', 'not identified')
df_Postushkov['total_postushkov'] = np.where(df_Postushkov['R'] > 1, 'stable', df_Postushkov['total_postushkov'])
```

```python
df_Postushkov['total_postushkov'].value_counts()
```

```python
df_Postushkov_total = df_Postushkov[['ID', 'total_postushkov']]
```

Добавляем к данным результат расчета по второй модели

```python
df_report_two_methods = pd.merge(df_report_first_method, df_Postushkov_total, how = 'left', on = 'ID')
df_report_two_methods
```

**R-модель, разработанная в Иркутской государственной экономической академии**, также используется для оценки вероятности банкротства.

$$
R = 8.38*K_1+K_2+0.054K_3+0.63K_4
$$

1. Коэффициент чистого оборотного капитала
2. Рентабельность собственного капитала
3. Рентабельность активов
4. Рентабельность затрат

R от 0 до 0,18 - высокая вероятность банкротства, R от 0,18 до 0,32 - средняя.
R от 0,32 до 0,42 - низкая вероятность банкротства, больше 0,42 минимальная.

```python
df_Irkutsk = pd.DataFrame({
    'ID': df_report_not_identified['ID'],
    'working_capital _ratio': (df_report_not_identified['1200'] - df_report_not_identified['1500']) / df_report_not_identified['1600'],
    'ROE_ratio': df_report_not_identified['2400'] / df_report_not_identified['1300'],
    'ROA_ratio': df_report_not_identified['2400'] / df_report_not_identified['1600'],
    'cost_margin_ratio': df_report_not_identified['2400'] / df_report_not_identified['2120']
})
```

```python
df_Irkutsk['R'] = 8.38 * df_Irkutsk['working_capital _ratio'] + df_Irkutsk['ROE_ratio'] + 0.054 * df_Irkutsk['ROA_ratio'] + 0.63 * df_Irkutsk['cost_margin_ratio']
```

```python
df_Irkutsk['total_irkutsk'] = np.where(df_Irkutsk['R'] <= 0.32, 'unstable', 'stable')
```
